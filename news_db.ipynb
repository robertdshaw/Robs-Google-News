```python
def get_latest_published_date():
    """Execute a query to get the latest (maximum) publishedAt date"""
    conn = sqlite3.connect('news.db')
    cursor = conn.cursor()
    cursor.execute("SELECT MAX(publishedAt) FROM news")
    result = cursor.fetchone()[0]  
    conn.close()
    if result:
        return result
    else:
        return '1900-01-01'  
latest_published_at = get_latest_published_date()
print(f"The latest article was published on: {latest_published_at}")
```
